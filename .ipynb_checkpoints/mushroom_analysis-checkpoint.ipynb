```python
## Housekeeping

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import json
```

```python
## Import jsons and stick them in a dataframe

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### DEFINE FOLDER NAMES FOR SAVING OBJECTS AND FIGURES
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
# Initializing an empty dictionary
folderDict = {}

# Adding key-value pairs to the dictionary
folderDict['jsonData'] = '/Users/rb1725/Documents/Repositories/nivturk_broderson/data'
```

```python
# Replace 'your_folder_path' with the path to your folder containing JSON files
folder_path = folderDict['jsonData']


# Function to read JSON files and return a DataFrame with an added ID column
def read_json_files(folder_path):
    dataframes = []
    
    # List all files in the folder
    files = os.listdir(folder_path)
    
    for file_id, file in enumerate(files, start=1):
        if file.endswith('.json'):
            file_path = os.path.join(folder_path, file)
            
            # Read JSON file and convert to DataFrame
            with open(file_path, 'r') as json_file:
                json_data = json.load(json_file)
                df = pd.DataFrame(json_data)
                
                # Add an ID column with the file_id
                df['ID'] = file_id
                
                dataframes.append(df)
    
    # Concatenate all DataFrames into one
    result_df = pd.concat(dataframes, ignore_index=True)
    
    return result_df

# Call the function with the folder path
result_dataframe = read_json_files(folder_path)

# Display the resulting DataFrame
print(result_dataframe)
```
